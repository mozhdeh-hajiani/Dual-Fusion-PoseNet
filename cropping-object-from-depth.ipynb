```python
import os
import cv2
import yaml
from tqdm import tqdm

# ==== Paths ====
LINEMOD_PATH = '/content/Linemod_preprocessed/data'
YOLO_PATH = '/content/YOLO_LINEMOD'
IMAGE_WIDTH = 640
IMAGE_HEIGHT = 480

CLASS_NAMES = ['ape', 'benchvise', 'bowl', 'can', 'cat', 'cup', 'driller', 'duck',
               'glue', 'holepuncher', 'iron', 'lamp', 'phone', 'cam', 'eggbox']

# ==== Create output folders ====
for split in ['train', 'test']:
    os.makedirs(f'{YOLO_PATH}/images/{split}', exist_ok=True)
    os.makedirs(f'{YOLO_PATH}/depth/{split}', exist_ok=True)
    os.makedirs(f'{YOLO_PATH}/labels/{split}', exist_ok=True)

def convert_bbox_to_yolo(x, y, w, h, img_w, img_h):
    x_center = (x + w / 2) / img_w
    y_center = (y + h / 2) / img_h
    w_norm = w / img_w
    h_norm = h / img_h
    return x_center, y_center, w_norm, h_norm

# ==== Process each class ====
for obj_id, class_name in enumerate(CLASS_NAMES, start=1):
    class_folder = f'{obj_id:02d}'
    folder_path = os.path.join(LINEMOD_PATH, class_folder)

    train_txt_path = os.path.join(folder_path, 'train.txt')
    test_txt_path = os.path.join(folder_path, 'test.txt')
    gt_path = os.path.join(folder_path, 'gt.yml')

    if not os.path.exists(train_txt_path) or not os.path.exists(test_txt_path) or not os.path.exists(gt_path):
        print(f"Skipping {class_name} (missing files)")
        continue

    train_ids = [int(line.strip()) for line in open(train_txt_path)]
    test_ids = set(int(line.strip()) for line in open(test_txt_path))
    gt = yaml.safe_load(open(gt_path))

    for img_id_str in tqdm(gt.keys(), desc=f'Class {class_name}'):
        img_id = int(img_id_str)
        img_name = f'{img_id:04d}.png'
        rgb_path = os.path.join(folder_path, 'rgb', img_name)
        depth_path = os.path.join(folder_path, 'depth', img_name)

        if not os.path.exists(rgb_path) or not os.path.exists(depth_path):
            continue

        # Determine subset
        if img_id in train_ids:
            subset = 'train'
        elif img_id in test_ids:
            subset = 'test'
        else:
            continue

        base_name = f'{class_folder}_{img_name}'
        out_rgb_path = os.path.join(YOLO_PATH, f'images/{subset}', base_name)
        out_depth_path = os.path.join(YOLO_PATH, f'depth/{subset}', base_name)
        out_label_path = os.path.join(YOLO_PATH, f'labels/{subset}', base_name.replace('.png', '.txt'))

        # Save RGB image
        rgb_img = cv2.imread(rgb_path)
        cv2.imwrite(out_rgb_path, rgb_img)

        # Save depth map
        depth_img = cv2.imread(depth_path, cv2.IMREAD_UNCHANGED)
        cv2.imwrite(out_depth_path, depth_img)

        # Write YOLO label
        with open(out_label_path, 'w') as f:
            for obj in gt[img_id]:
                if obj['obj_id'] != obj_id:
                    continue
                x, y, w, h = obj['obj_bb']
                yolo_box = convert_bbox_to_yolo(x, y, w, h, rgb_img.shape[1], rgb_img.shape[0])
                label_line = f"{obj_id - 1} {' '.join(f'{v:.6f}' for v in yolo_box)}\n"
                f.write(label_line)

# ==== Create data.yaml ====
with open(os.path.join(YOLO_PATH, 'data.yaml'), 'w') as f:
    f.write(f'''train: {YOLO_PATH}/images/train
val: {YOLO_PATH}/images/train  # Reusing train as validation to satisfy YOLOv8
nc: {len(CLASS_NAMES)}
names: {CLASS_NAMES}
''')
```

```python
#train
import os
import cv2
import pandas as pd
from ultralytics import YOLO

def detect_crop_and_save_results(
    folder_path="/content/YOLO_LINEMOD/images/train",
    depth_folder="/content/YOLO_LINEMOD/depth/train",
    model_path="/content/drive/MyDrive/yolo_models/runs_backup/detect/linemod_yolo_v8n2/weights/best.onnx",
    imgsz=640,
    conf=0.25,
    save_dir="/content/drive/MyDrive/yolo_models/linemod_yolo_v8n/depth_crop/train/train_cropped_objects",
    csv_output="/content/drive/MyDrive/yolo_models/linemod_yolo_v8n/depth_crop/train/detection_results.csv"
):
    os.makedirs(save_dir, exist_ok=True)

    model = YOLO(model_path)
    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    detection_records = []

    for img_file in image_files:
        img_path = os.path.join(folder_path, img_file)
        orig_img = cv2.imread(img_path)
        if orig_img is None:
            print(f"Skipping {img_file}, image not found.")
            continue

        orig_h, orig_w = orig_img.shape[:2]

        # Get depth map path
        depth_img_path = os.path.join(depth_folder, img_file)
        depth_img = cv2.imread(depth_img_path, cv2.IMREAD_UNCHANGED)
        if depth_img is None:
            print(f"Skipping {img_file}, depth map not found.")
            continue

        # Run YOLOv8 inference on original RGB image
        results = model.predict(source=orig_img, imgsz=imgsz, conf=conf, verbose=False)

        for r in results:
            for i, box in enumerate(r.boxes.data):
                x1, y1, x2, y2 = map(int, box[:4])
                class_id = int(box[5])
                conf_score = float(box[4])

                # Scale box coords from imgsz to original image size
                scale_x = orig_w / imgsz
                scale_y = orig_h / imgsz
                x1_orig = int(x1 * scale_x)
                x2_orig = int(x2 * scale_x)
                y1_orig = int(y1 * scale_y)
                y2_orig = int(y2 * scale_y)

                # Clamp bounding box to image dimensions
                x1_orig = max(0, x1_orig)
                y1_orig = max(0, y1_orig)
                x2_orig = min(orig_w, x2_orig)
                y2_orig = min(orig_h, y2_orig)

                # Crop and save depth image
                base_name = os.path.splitext(img_file)[0]
                crop_name = f"{base_name}.png"  # index to avoid overwrite
                crop_path = os.path.join(save_dir, crop_name)
                cropped = depth_img[y1_orig:y2_orig, x1_orig:x2_orig]
                cv2.imwrite(crop_path, cropped)

                # Save detection info
                detection_records.append({
                    "filename": img_file,
                    "class_id": class_id,
                    "confidence": conf_score,
                    "x1": x1_orig,
                    "y1": y1_orig,
                    "x2": x2_orig,
                    "y2": y2_orig,
                    "cropped_image": crop_path
                })

                print(f" Saved: {crop_path}")

    # Save results as CSV
    df = pd.DataFrame(detection_records)
    df.to_csv(csv_output, index=False)
    print(f"\n📄 Detection results saved to: {csv_output}")

# Run it
detect_crop_and_save_results()
```

```python
import os
import cv2
import pandas as pd
from ultralytics import YOLO

def detect_and_crop_single_image(
    image_path,
    depth_path,
    model_path="/content/drive/MyDrive/yolo_models/runs_backup/detect/linemod_yolo_v8n2/weights/best.onnx",
    imgsz=640,
    conf=0.25,
    save_dir="/content/drive/MyDrive/yolo_models/linemod_yolo_v8n/depth_crop/test/test_cropped_objects",
    csv_output="/content/drive/MyDrive/yolo_models/linemod_yolo_v8n/depth_crop/test/detection_results_single.csv"
):
    os.makedirs(save_dir, exist_ok=True)

    model = YOLO(model_path)

    orig_img = cv2.imread(image_path)
    if orig_img is None:
        print(f"Image not found: {image_path}")
        return

    depth_img = cv2.imread(depth_path, cv2.IMREAD_UNCHANGED)
    if depth_img is None:
        print(f"Depth map not found: {depth_path}")
        return

    orig_h, orig_w = orig_img.shape[:2]

    # Run YOLOv8 inference
    results = model.predict(source=orig_img, imgsz=imgsz, conf=conf, verbose=False)

    detection_records = []

    for r in results:
        for i, box in enumerate(r.boxes.data):
            x1, y1, x2, y2 = map(int, box[:4])
            class_id = int(box[5])
            conf_score = float(box[4])

            # Scale box coords
            scale_x = orig_w / imgsz
            scale_y = orig_h / imgsz
            x1_orig = int(x1 * scale_x)
            x2_orig = int(x2 * scale_x)
            y1_orig = int(y1 * scale_y)
            y2_orig = int(y2 * scale_y)

            x1_orig = max(0, x1_orig)
            y1_orig = max(0, y1_orig)
            x2_orig = min(orig_w, x2_orig)
            y2_orig = min(orig_h, y2_orig)

            # Crop and save depth image
            base_name = os.path.splitext(os.path.basename(image_path))[0]
            crop_name = f"{base_name}_crop{i}.png"
            crop_path = os.path.join(save_dir, crop_name)
            cropped = depth_img[y1_orig:y2_orig, x1_orig:x2_orig]
            cv2.imwrite(crop_path, cropped)

            detection_records.append({
                "filename": os.path.basename(image_path),
                "class_id": class_id,
                "confidence": conf_score,
                "x1": x1_orig,
                "y1": y1_orig,
                "x2": x2_orig,
                "y2": y2_orig,
                "cropped_image": crop_path
            })

            print(f" Saved: {crop_path}")

    # Save CSV
    if detection_records:
        df = pd.DataFrame(detection_records)
        df.to_csv(csv_output, index=False)
        print(f"\n Detection result saved to: {csv_output}")
    else:
        print("No detections made.")

# Example usage:
detect_and_crop_single_image(
    image_path="/content/YOLO_LINEMOD/images/test/12_1094.png",
    depth_path="/content/YOLO_LINEMOD/depth/test/12_1094.png"
)
```
